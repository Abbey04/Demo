## Challenge 002 Time series 
Data is from Google finance. It is looking at the stock market opening and closing data from January 2020.

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates

import seaborn; seaborn.set()

from pandas_datareader import data
```

```python
import pandas_datareader.data as data

goog = data.DataReader('^DJI', 'stooq',start = '2020-01-01')
```

```python
goog['Year']=goog.index.year
goog['Month'] = goog.index.month
goog['Month Name'] = goog.index.month_name()
goog['Weekday Name'] = goog.index.day_name()
```

```python
goog.head(10)
```

```python
fig,ax = plt.subplots(5,1,figsize= (18,11))
plt.tight_layout()

ax[0].plot(goog.index, goog.Open,color='b',alpha =0.75)
ax[0].set_title('Open')

ax[1].plot(goog.index, goog.Close,color='g')
ax[1].set_title('Close')

ax[2].plot(goog.index, goog.High,color='y')
ax[2].set_title('High')

ax[3].plot(goog.index, goog.Low,color='c')
ax[3].set_title('Low')

ax[4].plot(goog.index, goog.Volume/1000000,color='m')
ax[4].set_title('Volume')
ax[4].set_ylabel('Volumne (M)')
```
